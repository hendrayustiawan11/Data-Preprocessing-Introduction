```python
bin_labels_5 = ['Bronze', 'Silver', 'Gold', 'Platinum']
results, bin_edges = pd.qcut(df['price'], q=4,
                             labels=bin_labels_5,
                             retbins=True)
```

```python
results
```

```python
results_table = pd.DataFrame(zip(bin_edges, bin_labels_5),
                             columns=['Threshold', 'Tier'])
```

```python
results_table
```

```python
df['results'] = results
```

```python
df.sort_values(by='price')
```

```python
df['results'].value_counts()
```

```python
bins = [0, 1000, 5000, 10000, 25000, 50000, 100000]
labels = [1,2,3,4,5,6]
df['binned_km'] = pd.cut(df['km'], bins=bins, labels=labels)
df
```
